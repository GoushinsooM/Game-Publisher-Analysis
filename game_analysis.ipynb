```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/vgsales.csv')
df['Year'] = df['Year'].astype('Int64')
df.isnull().sum()
```

# Region sales per publisher

```python
total_sales_publisher_na = df.groupby('Publisher')['NA_Sales'].sum().sort_values(ascending=False)
total_sales_publisher_na.head(30).plot.barh(title='Publisher sales on NA')
plt.xlabel('NA_Sales')
plt.ylabel('Publisher')
```

```python
total_sales_publisher_eu = df.groupby('Publisher')['EU_Sales'].sum().sort_values(ascending=False)
total_sales_publisher_eu.head(30).plot.barh(title='Publisher sales on EU')
plt.xlabel('EU_Sales')
plt.ylabel('Publisher')
```

```python
total_sales_publisher_jp = df.groupby('Publisher')['JP_Sales'].sum().sort_values(ascending=False)
total_sales_publisher_jp.head(30).plot.barh(title='Publisher sales on JP')
plt.xlabel('JP_Sales')
plt.ylabel('Publisher')
```

```python
total_sales_publisher_other = df.groupby('Publisher')['Other_Sales'].sum().sort_values(ascending=False)
total_sales_publisher_other.head(30).plot.barh(title='Publisher sales on other region')
plt.xlabel('Other_Sales')
plt.ylabel('Publisher')
```

# Global sales per publisher

```python
total_sales_publisher_global = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False)
total_sales_publisher_global.head(30).plot.barh(title='Global publisher sales')
plt.xlabel('Global_Sales')
plt.ylabel('Publisher')
```

# Region sales per genre

```python
na_total_sales_gender = df.groupby('Genre')['NA_Sales'].sum().sort_values(ascending=False)
na_total_sales_gender.head(12).plot.bar(title='NA sales per game genre')
plt.xlabel('Genre')
plt.ylabel('NA_Sales')
```

```python
eu_total_sales_gender = df.groupby('Genre')['EU_Sales'].sum().sort_values(ascending=False)
eu_total_sales_gender.head(12).plot.bar(title='EU sales per game genre')
plt.xlabel('Genre')
plt.ylabel('EU_Sales')
```

```python
jp_total_sales_gender = df.groupby('Genre')['JP_Sales'].sum().sort_values(ascending=False)
jp_total_sales_gender.head(12).plot.bar(title='JP sales per game genre')
plt.xlabel('Genre')
plt.ylabel('JP_Sales')
```

```python
other_total_sales_gender = df.groupby('Genre')['Other_Sales'].sum().sort_values(ascending=False)
other_total_sales_gender.head(12).plot.bar(title='Other sales per game genre')
plt.xlabel('Genre')
plt.ylabel('Other_Sales')
```

# Global total sales per genre

```python
global_total_sales_gender = df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)
global_total_sales_gender.head(12).plot.bar(title='Global sales per game genre')
plt.xlabel('Genre')
plt.ylabel('Global_Sales')
```

# Sales over the years 

```python
over_years_na = df.groupby('Year')['NA_Sales'].count()
over_years_na.plot.bar(title='NA sales over the years')
plt.xlabel('Year')
plt.ylabel('Sales')
```
